```python
##Outlier Detection and Removing using Z-Score
```

```python
sns.boxplot('duration'  , data =df1)
```

```python
sns.boxplot('age'  , data =df1)
```

```python
print(df1.duration.mean())
print(df1.age.mean())
```

```python
print(df1.duration.std())
print(df1.age.std())
```

```python
df1['zscore1'] = ( df1.duration - df1.duration.mean() ) / df1.duration.std()
df1['zscore2'] = ( df1.age - df1.age.mean() ) / df1.age.std()
df1
```

```python
df1[(df1.zscore1<-3) | (df1.zscore1>3)|(df1.zscore2<-3) | (df1.zscore2>3)]
```

```python
df_no_outlier = df1[(df1.zscore1<-3) | (df1.zscore1>3)]
df_no_outlier
```

```python
## Outlier Removal using IQR method 
```

```python
iqrnum=['age','duration']
for col in iqrnum:
    Q1=df1[col].quantile(0.25)
    Q3=df1[col].quantile(0.75)
    IQR= Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    filter=(df1[col] >= lower_limit)&(df1[col] <= upper_limit)
    df1=df1.loc[filter]
    
df1.shape

```
